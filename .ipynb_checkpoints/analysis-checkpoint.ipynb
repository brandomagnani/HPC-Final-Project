```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 16
```

```python
path = '/home/aeb783/HPC-Final-Project/'
```

Read sgd datasets

```python
n_threads = 8
# thread_counts = np.arange(1, n_threads+1)
thread_counts = 2**np.arange(6, dtype=int)
thread_counts
```

```python
dfs = {}
for i in thread_counts:
    df = (
        pd.read_csv(path + 'output/1e6/{}_threads_sgd.csv'.format(i))
        .iloc[0:-1]
        .astype(float)
    )
    dfs['{}_threads'.format(i)] = df
```

```python
print('2 threads:')
print(dfs['2_threads'].tail())
print('\n32 threads:')
print(dfs['32_threads'].tail())
```

## Residual norm vs iteration

```python
fig, ax = plt.subplots(figsize=(12,8))

for i in thread_counts:
    df = dfs['{}_threads'.format(i)]
    
    ax.semilogy(df['residual_norm'], label='n_threads={}'.format(i))
    
ax.set(
    xlabel='Iteration',
    ylabel='Residual norm',
    title='Residual norm by iteration',
)

ax.legend();
```

## Residual norm by time

```python
fig, ax = plt.subplots(figsize=(12,8))

for i in thread_counts:
    df = dfs['{}_threads'.format(i)]
    x = df['time']
    y = df['residual_norm']
    
    ax.semilogy(x, y, label='n_threads={}'.format(i))
    
ax.set(
    xlabel='Time (s)',
    ylabel='Residual norm',
    title='Residual norm vs time',
    xlim=(-10,1210)
)

ax.legend();
```

## Time for convergence

```python
df = dfs['8_threads']
df[df['residual_norm'] / df['residual_norm'].iloc[0] < 0.01]
```

```python
def convergence_time(df, tol=0.01):
    """Computes time to convergence"""
    if len(df[df['residual_norm'] / df['residual_norm'].iloc[0] < tol]) == 0:
        return np.nan
    else:
        conv_time = df[df['residual_norm'] / df['residual_norm'].iloc[0] < tol]['time'].iloc[0]
        return conv_time
```

```python
convergence_times = np.empty_like(thread_counts, dtype=float)

for i, n_threads in enumerate(thread_counts):
    df = dfs['{}_threads'.format(n_threads)]
    convergence_times[i] = convergence_time(df)
```

```python
convergence_times
```

```python
fig, ax = plt.subplots()

bs = [4, 5, 6, 7]

for b in bs:
    print(b)
    
    dfs = {}
    for i in thread_counts:
        df = (
            pd.read_csv(path + 'output/1e{}/{}_threads_sgd.csv'.format(b, i))
            .iloc[0:-1]
            .astype(float)
        )
        dfs['{}_threads'.format(i)] = df
    
    convergence_times = np.empty_like(thread_counts, dtype=float)

    for i, n_threads in enumerate(thread_counts):
        df = dfs['{}_threads'.format(n_threads)]
        convergence_times[i] = convergence_time(df)

    ax.plot(thread_counts, convergence_times, label=r'$n=10^{}$'.format(b),
                ls='-', marker='x');
    
ax.set_xscale('log', base=2)
ax.set_yscale('log', base=10)

ax.set(
    title=r'Time to convergence (tol=0.01)',
    xlabel='Number of threads',
    ylabel='Simulation time (s)',
    ylim=(1, 1e4),
)

ax.legend(bbox_to_anchor=(1,0.75));
```

## Flop rate

```python
dfs = {}
for i in thread_counts:
    df = (
        pd.read_csv(path + 'output/1e6/{}_threads_sgd.csv'.format(i))
        .iloc[0:-1]
        .astype(float)
    )
    dfs['{}_threads'.format(i)] = df
    
flop_rates = np.empty_like(thread_counts)

for i, n in enumerate(thread_counts):
    df = dfs['{}_threads'.format(n)]
    flop_rates[i] = df['flops'].mean()
```

```python
fig, ax = plt.subplots()

for b in [6]:
    dfs = {}
    for i in thread_counts:
        df = (
            pd.read_csv(path + 'output/1e{}/{}_threads_sgd.csv'.format(b, i))
            .iloc[0:-1]
            .astype(float)
        )
        dfs['{}_threads'.format(i)] = df
    
    flop_rates = np.empty_like(thread_counts, dtype=float)

    for i, n in enumerate(thread_counts):
        df = dfs['{}_threads'.format(n)]
        flop_rates[i] = df['flops'].mean()
    ax.plot(thread_counts, flop_rates, ls='-', marker='x', label='$n=10^{}'.format(b))

ax.set_xscale('log', base=2)
ax.set(
    title='Bandwidth vs thread count',
    xticks=thread_counts,
    xlabel='Number of threads',
    ylabel='GFlop/s',
    ylim=(0, None)
);
# ax.legend();
```
